```python
import matplotlib.pyplot as plt
import csv

def check_sration(filename, year):
    f = open(filename, 'r', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)
    
    male = []
    female = []
    diff = []
    
    for row in data:
        if '  (5000000000)' in row[0]:
            for i in row[3:104]:
                male.append(-int(i))
            for i in row[106:]:
                female.append(int(i))

    for j in range(101):
        diff.append(male[j]+female[j])
    
    f.close()
    
    #subplot 
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
    
    axes[0].pie([abs(sum(male)),sum(female)], labels=['male', 'female'], colors=['cornflowerblue', 'crimson'], autopct='%.3f%%')
    axes[0].set_title('Jeju Island Population Ratio by Gender in '+str(year))
    
    axes[1].barh(range(101), male, label='Male')
    axes[1].barh(range(101), female, label='Female', color='crimson')
    axes[1].barh(range(101), diff, label='Difference', color='plum')
    axes[1].legend()
    axes[1].set_ylabel("Age")
    axes[1].set_xlabel("Population(absolute value)")
    axes[1].set_title('Jeju Island Population by Gender in '+ str(year))
    
    plt.show()
    
def main():
    check_sration('jeju2008pop.csv', 2008)
    check_sration('jeju2012pop.csv', 2012)
    check_sration('jeju2016pop.csv', 2016)
    check_sration('jeju2020pop.csv', 2020)
    check_sration('jeju2022pop.csv', 2022)
    

if __name__ == '__main__':
    main()
```
