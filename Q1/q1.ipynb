Average monthly temperatures by region in 2022

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

def monthly_temp(filename, labelname) :
    f=open(filename, 'r', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)
         
    month =[[], [], [], [], [], [] ,[],[],[],[],[],[]]

    for row in data:
        if row[-1] != '':
            month[int(row[0].split('-')[1])-1].append(float(row[2]))        
    
    amonth = []
    for i in range(12):
        amonth.append(sum(month[i])/len(month[i]))
    
    f.close()
    plt.plot(amonth, label=labelname)

def main():
    monthly_temp('nationwide.csv', 'Nationwide')
    monthly_temp('seoul.csv', 'Seoul')
    monthly_temp('busan.csv', 'Busan')
    monthly_temp('daejeon.csv', 'Daejeon')
    monthly_temp('jeju.csv', 'Jeju')
    plt.title('Average monthly temperatures by region in 2022')
    plt.xlabel("months")
    plt.xticks(np.arange(0,12,1), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dev'])
    plt.ylabel("Celsius")
    plt.yticks([-5, 0, 5, 10, 15, 20, 25, 30])
    plt.legend()
    plt.show()
    
    
if __name__ == '__main__':
    main()
```
